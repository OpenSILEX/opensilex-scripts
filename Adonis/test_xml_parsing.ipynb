```python
def parsePath(s, root):

    elements = s.split(" ")
    l2 = []

    for e in elements:

        split_path = e.split("/")
        tmp = root

        for path_e in split_path[1:]:
            try:
                path_e = int(path_e)
                tmp = tmp.findall("*")[path_e]
            except:
                if '.' in path_e:
                    l = path_e[1:].split(".")
                    tmp = tmp.findall("./" + l[0])[int(l[1])]
                else:
                    tmp = tmp.findall("./" + path_e[1:])[0]
        l2.append(tmp)
    return(l2)
```

```python
d = {}
i = 0
for elem in root.iter():
    i += 1
    if elem.tag not in d.keys():
        d[elem.tag] = pd.DataFrame()
    
    newrow = elem.attrib.copy()
    for k in newrow.keys():
        if "/" in newrow[k]:
            newrow[k] = [x.attrib for x in parsePath(newrow[k], root)]

    d[elem.tag] = d[elem.tag].append(newrow, ignore_index=True)
```

```python
def extract(elem, d, parents, n):
    if elem.tag not in d.keys():
        d[elem.tag] = pd.DataFrame()
    
    newrow = elem.attrib.copy()

    newrow.update(parents)

    d[elem.tag] = d[elem.tag].append(newrow, ignore_index=True)

    parents.update({elem.tag : n})

    n = 0
    for child in elem:
        extract(child, d, parents.copy(), n)
        n += 1
```
