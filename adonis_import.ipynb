```python
# FACTOR

def createFactorByExperiment(expe, facteurs, factor_category, factor_description, pythonClient):
    facteur_api = opensilexClientTools.FactorsApi(pythonClient)
    exp_fac_cat = facteur_api.search_categories()['result']
    fac_cat = []
    for type in exp_fac_cat:
        fac_cat.append(type.uri)
    try:
        factor_category in fac_cat
    except Exception as e:
        print("Exception : %s\n" % e)
        return None

    set_facteurs = list(set(facteurs["nom"]))
    for fac in set_facteurs:
        one_facteur = facteurs[facteurs["nom"] == fac]
        facteur_mod = []
        for mod in range(0, len(one_facteur)):
            facteur_mod.append(one_facteur["modalites"][mod].get("@valeur"))
        facteurDTO = opensilexClientTools.FactorCreationDTO(
                                name = fac, 
                                category = factor_category, 
                                description = factor_description,  
                                levels={"name":facteur_mod, "description": facteur_mod}, 
                                experiment = expe)
                                # exact_match="", close_match="", broad_match="", narrow_match="")  potentiels liens à intégrer
        create_facteur = facteur_api.create_factor(body=facteurDTO)
```
